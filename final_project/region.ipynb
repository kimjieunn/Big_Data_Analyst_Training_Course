```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 문화공간

```python
culture = pd.read_csv('./data/문화공간.csv',encoding='cp949')
culture.head(5)
```

```python
# 필요없다고 생각하는 변수 제거
# 번호, 팩스번호, 홈페이지, 개관일자, 대표이미지, YELLOW, GREEN, BLUE, RED, 공항버스
culture.drop(['번호','팩스번호','홈페이지','개관일자','객석수','대표이미지','기타사항','지하철','버스정거장','YELLOW', 'GREEN', 'BLUE', 'RED', '공항버스'],inplace=True,axis=1)
culture
```

```python
culture.isnull().sum()
```

```python
culture.info()
```

```python
culture['주제분류'].value_counts()
```

```python
# '경기도' 값을 제외한 값만 저장
culture = culture[culture['주소'].str.contains('서울|노원|한강|종로|일원|망우|자양|강남|구로|홍릉|서대문구')].reset_index()
culture.drop(['index'],inplace=True,axis=1)
culture.head()
```

```python
# !pip install folium
```

```python
import folium as g
```

```python
culture_map = g.Map(location=[37.55,126.98],
                   zoom_start=12)
culture_map
```

```python
# 서울 문화공간만 지도에 표시
for i in range(len(culture)):
    marker01 = g.Marker([culture.loc[i]['위도'],culture.loc[i]['경도']],
                       popup = culture.loc[i]['문화시설명'],
                       icon = g.Icon(color='blue'))
    marker01.add_to(culture_map)
culture_map
```

```python
culture.isnull().sum()
```

```python
culture.drop(['전화번호','관람시간','관람료','휴관일','무료구분'],inplace=True,axis=1)
culture.isnull().sum()
```

```python
# culture.to_csv('culture.csv',index=False, encoding="utf-8-sig")
```

## 공원

```python
park = pd.read_csv('./data/공원.csv',encoding='cp949')
park.head(5)
```

```python
# 필요없다고 생각이 드는 변수 삭제
park.drop(['연번','면적','주요식물','안내도','오시는길','이용시참고사항','이미지','바로가기'],inplace=True,axis=1)
park.head(5)
```

```python
# '경기도 과천시' 값을 제외한 값만 저장
s_park = park[park['공원주소'].str.contains('서울|노원')].reset_index()
s_park.drop(['index'],inplace=True,axis=1)
s_park.head()
```

```python
s_park.info()
```

```python
park_map = g.Map(location=[37.55,126.98],
                   zoom_start=12)
park_map
```

```python
# 서울 공원만 지도에 표시
for i in range(len(s_park)):
    marker01 = g.Marker([s_park.loc[i]['Y좌표(WGS84)'],s_park.loc[i]['X좌표(WGS84)']],
                       popup = s_park.loc[i]['공원명'],
                       icon = g.Icon(color='blue'))
    marker01.add_to(park_map)
park_map
```

```python
s_park.isnull().sum()
```

```python
s_park.drop(['공원개요','개원일','주요시설','X좌표(GRS80TM)','Y좌표(GRS80TM)'],inplace=True,axis=1)
s_park.isnull().sum()
```

```python
s_park.to_csv('park.csv',index=False, encoding="cp949")
```

## 마트

```python
store = pd.read_csv('./data/대규모점포.csv',encoding='cp949')
store.head()
```

```python
store['상세영업상태명'].value_counts()
```

```python
store.isnull().sum()
```

```python
store.drop(['번호','개방서비스아이디','개방자치단체코드','인허가취소일자','관리번호','인허가일자','최종수정시점','데이터갱신구분','데이터갱신일자','영업상태구분코드','상세영업상태코드','소재지전화','소재지면적','소재지우편번호','도로명우편번호'],inplace=True,axis=1)
store.head()
```

```python
s_store = store[store['소재지전체주소'].str.contains('서울',na=False)].reset_index()
s_store.drop(['index'],inplace=True,axis=1)
s_store
```

```python
s_store['상세영업상태명'].value_counts()
```

```python
s_store = s_store[s_store['상세영업상태명'].str.contains('정상영업',na=False)].reset_index()
s_store.drop(['index'],inplace=True,axis=1)
s_store
```

```python
s_store.isnull().sum()
```

```python
s_store.groupby(s_store['영업상태명'])['영업상태명'].count()
```

```python
s_store.drop(['폐업일자','휴업시작일자','휴업종료일자','재개업일자','도로명전체주소'],inplace=True,axis=1)
s_store.head()
```

```python
import folium as g
store_map = g.Map(location=[37.55,126.98],
                   zoom_start=12)
store_map
```

```python
# 서울 대규모점포 지도에 표시
for i in range(len(s_store)):
    marker01 = g.Marker([s_store.loc[i]['위도'],s_store.loc[i]['경도']],
                       popup = s_store.loc[i]['사업장명'],
                       icon = g.Icon(color='blue'))
    marker01.add_to(store_map)
store_map
```

```python
s_store.isnull().sum()
```

```python
s_store.drop(['개방서비스명','점포구분명','영업상태명'],inplace=True,axis=1)
s_store.isnull().sum()
```
